```python
def linmodl(m,b,x):
    y = m * x + b
    return y

def lnL(theta,x,y,yerr):
    m, b = theta
    modl = linmodl(m,b,x)
    inv_sig2 = np.reciprocal(np.square(yerr))
    return -0.5 * np.sum(np.multiply(np.square((y - modl)),inv_sig2) - np.log(inv_sig2))

def lnP(theta, *args):
    m, b = theta
    if m > 1.0 or m <-1.0:
        return -np.inf
    elif b < -5.0 or b > 5.0:
        return -np.inf
    return lnL(theta, *args)
```

```python
from scipy.optimize import minimize

slopes = []
inters = []
for i in range(len(elems)-5):
    cln = np.where((r13[elems[i]]>-100)&(r13['SNR']>=100))
    clnr = np.where((r13[elems[i]]>-3.0*gfunc.mad(r13[cln][elems[i]])+np.median(r13[cln][elems[i]]))&
                    (r13[elems[i]]<3.0*gfunc.mad(r13[cln][elems[i]])+np.median(r13[cln][elems[i]])))

    min_r = np.where(radius[clnr]<=0.5)

    m_guess = 0.0 
    b_guess = np.median(r13[elems[0]][clnr][min_r])

    exog = radius[clnr]
    endog = r13[elems[i]][clnr]
    endog_err = r13[elems_err[i]][clnr]

    nll = lambda *args: -lnL(*args)
    guess = np.array([m_guess, b_guess])
    soln = minimize(nll, guess, args=(exog, endog, endog_err))
    m_ml, b_ml = soln.x

    slopes.append(m_ml)
    inters.append(b_ml)
```
